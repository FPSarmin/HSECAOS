```python
# Сортировка выбором (Selection sort)
a = list()
n = len(a)
for i in range(n):
    min_pos = i
    for j in range(i, n):
        if a[j] < a[min_pos]:
            min_pos = j            
    (a[i], a[min_pos]) = (a[min_pos], a[i])
```

```python
# Сортировка пузырьком
# a[i] < a[i+1]
for i in range(n - 1):
    is_sorted = True
    for j in range(n - 1 - i):
        if a[j] > a[j + 1]:
            (a[j], a[j + 1]) = (a[j + 1], a[j]) 
            is_sorted = False
    if is_sorted:
        break
# [...... MAX .......]
# [.............  MAX]
```

```python
Быстрой сортировки (Quicksort). Алгоритм Хоара

# [.........] < [.........]
# [4, 1, 3, 2] < [7, 6, 8, 9]
# [1, 7, 4, 5, 3, 2, 6, 8]

# median = 5
# less = [1, 4, 3, 2]
# greater = [7, 5, 6, 8]
# result = quicksort(less) + quicksort(right)

# возьмём в качестве median — a[0]
# [1, 2, 3, 4, 5, 6, 7, 8]



def qsort(a):
    if (len(a) <= 1):
        return a
    median = a[rand.randrange(0, len(a))] # rand.choice(a)    
    less = []
    equal = []
    greater = []
    for x in a:
        if x < median:
            less.append(x)
        elif x > median:
            greater.append(x)
        else:
            equal.append(x)
    return qsort(less) + equal + qsort(greater)

# [..........][.............][............]
#      1/3          1/3           1/3


# хорошее разбиение, если len(less) и len(greater) <= 2/3 len(a)
log_{3/2}(n) — глубина хороших разбиений

# плохое разбиение, если оно не хорошее
2 * хороших разбиений 

# 3 * log_{3/2}(n)

# n * 3 log_{3/2}(n)  = O (n * log n)

def qsort(a, L, R):
    # хотим отсортировать a[L:R]
    if (R - L <= 1):
        return
    #if (R - L <= 10):
    #    запустим сортировку пузырьком
    median = a[rand.randrange(L, R)]    
    #[......... L............R ........]
    #[......... [...<median][....>=median]..... .........]
    #           L           m             i  R
    m = L
    #[][] ............
    # a[L] < median
    #[.][] ...........
    # m += 1
    # a[i] >= median
    #[.][.] ..........
    
    
    for i in range(L, R):
        if a[i] < median:
            (a[i], a[m]) = (a[m], a[i])
            m += 1
    #[.........[....][....]..........]
    #          L     m     R
    
    # [<median][ >= median]
    
    # [<median][=median][>median]
    m2 = m
    for i in range(m, R):
        if a[i] = median:
            (a[i], a[m]) = (a[m], a[i])
            m2 += 1
    
    qsort(a, L, m)
    qsort(a, m2, R)
    
    
```
